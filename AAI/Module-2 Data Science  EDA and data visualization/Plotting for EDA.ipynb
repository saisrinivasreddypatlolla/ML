```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
df.shape# how many data points and features
```

```python
df.columns # what are the column names in our dataset
```

```python
#how many data points for each class are present
#How many flowers for each spicies are present
df["Species"].value_counts()
#balanced-dataset vs imbalances datasets
#iris is a balanced dataset
```

```python
df.plot(kind="scatter",x="SepalLengthCm",y="SepalWidthCm")
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue="Species",size=4) \
.map(plt.scatter,"SepalLengthCm","SepalWidthCm")\
.add_legend();
plt.show();
```

```python
sns.FacetGrid(df,hue="Species",size=5) \
.map(sns.distplot,"PetalLengthCm") \
.add_legend();
plt.show()
```

```python
sns.boxplot(x="Species",y="PetalLengthCm",data=df)
plt.show()
```

```python
sns.violinplot(x="Species",y="PetalLengthCm",data=df,size=8)
plt.show()
```

```python
sns.jointplot(x="PetalLengthCm",y="PetalWidthCm",data=df)
```
