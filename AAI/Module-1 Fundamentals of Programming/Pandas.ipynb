# Concat data frames

```python
india_weathr = pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weathr
```

```python
us_weather = pd.DataFrame({
    "city":["NYC","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,35,75]
})
us_weather
```

```python
df = pd.concat([india_weathr,us_weather])
df
```

```python
df = pd.concat([india_weathr,us_weather],axis=1)
df
```

```python
df = pd.concat([india_weathr,us_weather],ignore_index=True)
df
```

```python
df.city
```

# Merge DataFrames

```python
temperature_df = pd.DataFrame({
    "city":["delhi","mumbai","bangalore","Hyderabad"],
    "temparature":[32,45,30,40]
})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city":["delhi","mumbai","bangalore"],
    "temparature":[65,68,75]
})
humidity_df
```

```python
df = pd.merge(temperature_df,humidity_df,on='city')
df
```

```python
df = pd.merge(temperature_df,humidity_df,on='city',how='outer')
df
```

```python
df.loc[:2]
```
